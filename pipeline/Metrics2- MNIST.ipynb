```python
def plot_methods(X, methods, file_name):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.bar(X, methods)
    ax.get_xticklabels()
    plt.xticks(rotation=30, ha='right')
    plt.ylabel("Score")
    plt.tight_layout()
    plt.show()
    fig.savefig(save_pth+file_name+file_ext)
```

```python
def box_plot(X, data, file_name, y_label='score', q='INT8'):
    
    xx = y_label
    
    condition = copy.deepcopy(X)
    for i in range(1, len(condition), 2):
        condition[i] = condition[i-1]
        
    a = ['FP32', q]
    b = int(len(X)/2)
    a = [a[i % len(a)] for i in range(len(a) * b)]    
        
    df_dict = {
        xx: np.array([]),
        "Method":np.array([]),
        "Model":np.array([])
    }
    df = pd.DataFrame(df_dict)
    
    for i in range(int(len(data))):
        results_dict = {
            xx: data[i],
            "Method": condition[i],
            "Model":a[i]
            }
        results_df = pd.DataFrame(results_dict)
        df = pd.concat([df, results_df], ignore_index=True)
        
    ax = sns.boxplot(data=df, x="Method", y=y_label, hue="Model")
      
    
    ax.set_xlabel('')
    plt.savefig(save_pth+file_name+file_ext)
```

```python
def gen_table(table, ce, ce_q, edl_log, edl_log_q, edl_digamma, edl_digamma_q, duq, duq_q, scale=100):
    mean_ce            = np.mean(ce)*scale
    std_ce             = np.std(ce)*scale

    mean_ce_q          = np.mean(ce_q)*scale
    std_ce_q           = np.std(ce_q)*scale

    mean_edl_log       = np.mean(edl_log)*scale
    std_edl_log        = np.std(edl_log)*scale

    mean_edl_log_q     = np.mean(edl_log_q)*scale
    std_edl_log_q      = np.std(edl_log_q)*scale

    mean_edl_digamma   = np.mean(edl_digamma)*scale
    std_edl_digamma    = np.std(edl_digamma)*scale

    mean_edl_digamma_q = np.mean(edl_digamma_q)*scale
    std_edl_digamma_q  = np.std(edl_digamma_q)*scale

    mean_duq           = np.mean(duq)*scale
    std_duq            = np.std(duq)*scale

    mean_duq_q         = np.mean(duq_q)*scale
    std_duq_q          = np.std(duq_q)*scale

    s_ce      = str(round(mean_ce, 2))+" ± "+str(round(std_ce, 2))
    q_ce      = str(round(mean_ce_q, 2))+" ± "+str(round(std_ce_q, 2))
    s_edl_log = str(round(mean_edl_log, 2))+" ± "+str(round(std_edl_log, 2))
    q_edl_log = str(round(mean_edl_log_q, 2))+" ± "+str(round(std_edl_log_q, 2))
    s_edl_dig = str(round(mean_edl_digamma, 2))+" ± "+str(round(std_edl_digamma, 2))
    q_edl_dig = str(round(mean_edl_digamma_q, 2))+" ± "+str(round(std_edl_digamma_q, 2))
    s_duq     = str(round(mean_duq, 2))+" ± "+str(round(std_duq, 2))
    q_duq     = str(round(mean_duq_q, 2))+" ± "+str(round(std_duq_q, 2))
    
    diff_ce = str(round(abs(mean_ce - mean_ce_q),2))
    diff_edl_log = str(round(abs(mean_edl_log - mean_edl_log_q),2))
    diff_edl_dig = str(round(abs(mean_edl_digamma - mean_edl_digamma_q),2))
    diff_duq = str(round(abs(mean_duq - mean_duq_q),2))


    if "MNIST" in file_ext:
        a1  = "\\begin{table}[h!] \centering \setstretch{1.3} \\begin{tabular}{|c|l|c|c|c|}"
        a2  = "\hline \\textbf{Dataset} & \\textbf{Method} & \\textbf{FP32 (Standard)} & \\textbf{INT8 (Quantised)} & \\textbf{$\\Delta$ Mean} \\\ \hline \multirow{5}{*}{MNIST}"
        a3  = " & Cross Entropy &" +str(s_ce)+ "&" +str(q_ce)+ "&" +str(diff_ce)+ "\\\ \cline{2-5}"
        a4  = " & Evidential Log Likelhood &" +str(s_edl_log)+ "&" +str(q_edl_log)+ "&" +str(diff_edl_log)+"  \\\ \cline{2-5} "
        a5  = " & Evidential Digamma &" +str(s_edl_dig)+ "  & " +str(q_edl_dig)+ "&" +str(diff_edl_dig)+ " \\\ \cline{2-5} "
        a6  = " & DUQ & " +str(s_duq)+ " & " +str(q_duq)+ "&" +str(diff_duq)+ " \\\ \hline "

    if "CIFAR10" in file_ext:
        a1  = " \multirow{5}{*}{CIFAR10}"
        a2  = " & Cross Entropy & " +str(s_ce)+ " & " +str(q_ce)+ "&" +str(diff_ce)+" \\\ \cline{2-5} "
        a3  = " & Evidential Log Likelihood & " +str(s_edl_log)+ " & " +str(q_edl_log)+ "&" +str(diff_edl_log)+" \\\ \cline{2-5} "
        a4  = " & Evidential Digamma & " +str(s_edl_dig)+ " & " +str(q_edl_dig)+ "&" +str(diff_edl_dig)+" \\\ \cline{2-5} "
        a5  = " & DUQ & " +str(s_duq)+ " & " +str(q_duq)+ "&" +str(diff_duq)+" \\\ \hline "
        a6  = " \end{tabular} \caption{"+ table +"} \label{table: "+ table +"} \end{table}"

    a = a1 + a2 + a3 + a4 + a5 + a6
    print(a)
```

```python
sns.set(style="whitegrid")
import matplotlib.collections as clt
import ptitprince as pt

def rain_plot(df, file_name):
    dy="entropy"; dx="group"; ort="v";
    pal = ['#66c2a5', '#66c2a5', '#fc8d62', '#fc8d62']
    f, ax = plt.subplots(figsize=(7, 5))

    ax = pt.half_violinplot(x = dx, y = dy, data = df, palette = pal, bw = .2, cut = 0.,
    scale = "area", width = .6, inner = None, orient = ort)

    ax = sns.stripplot(x = dx, y = dy, data = df, palette = pal, edgecolor = "white",
    size = 3, jitter = 1, zorder = 0, orient = ort)

    ax = sns.boxplot(x = dx, y = dy, data = df, color = "black", width = .15, zorder = 10,\
    showcaps = True, boxprops = {'facecolor':'none', "zorder":10},\
    showfliers=True, whiskerprops = {'linewidth':2, "zorder":10},\
    saturation = 1, orient = ort)

    ax.set_xlabel('')
    #plt.xticks(rotation=30, ha='right')
    plt.savefig(save_pth+file_name+file_ext)
    plt.show()
```
