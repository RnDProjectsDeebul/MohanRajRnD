```python
def plot_methods(X, methods, file_name):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.bar(X, methods)
    ax.get_xticklabels()
    plt.xticks(rotation=30, ha='right')
    plt.ylabel("Score")
    plt.tight_layout()
    plt.show()
    fig.savefig(save_pth+file_name+file_ext)
```

```python
def box_plot(X, data, file_name, y_label='score'):
    
    condition = copy.deepcopy(X)
    for i in range(1, len(condition), 2):
        condition[i] = condition[i-1]
        
    a = ['FP32', 'INT8']
    b = int(len(X)/2)
    a = [a[i % len(a)] for i in range(len(a) * b)]    
        
    df_dict = {
        "Score":np.array([]),
        "Method":np.array([]),
        "Model":np.array([])
    }
    df = pd.DataFrame(df_dict)
    
    for i in range(int(len(data))):
        results_dict = {
            "Score": data[i],
            "Method": condition[i],
            "Model":a[i]
            }
        results_df = pd.DataFrame(results_dict)
        df = pd.concat([df, results_df], ignore_index=True)
        
    ax = sns.boxplot(data=df, x="Method", y="Score", hue="Model")
    ax.set_xlabel('')
    plt.savefig(save_pth+file_name+file_ext)
```

```python
sns.set(style="whitegrid")
import matplotlib.collections as clt
import ptitprince as pt

def rain_plot(df, file_name):
    dy="entropy"; dx="group"; ort="v";
    pal = ['#66c2a5', '#66c2a5', '#fc8d62', '#fc8d62']
    f, ax = plt.subplots(figsize=(7, 5))

    ax = pt.half_violinplot(x = dx, y = dy, data = df, palette = pal, bw = .2, cut = 0.,
    scale = "area", width = .6, inner = None, orient = ort)

    ax = sns.stripplot(x = dx, y = dy, data = df, palette = pal, edgecolor = "white",
    size = 3, jitter = 1, zorder = 0, orient = ort)

    ax = sns.boxplot(x = dx, y = dy, data = df, color = "black", width = .15, zorder = 10,\
    showcaps = True, boxprops = {'facecolor':'none', "zorder":10},\
    showfliers=True, whiskerprops = {'linewidth':2, "zorder":10},\
    saturation = 1, orient = ort)

    ax.set_xlabel('')
    #plt.xticks(rotation=30, ha='right')
    plt.savefig(save_pth+file_name+file_ext)
    plt.show()
```
