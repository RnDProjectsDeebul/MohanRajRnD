```python
def plot_methods(X, methods, file_name):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.bar(X, methods)
    ax.get_xticklabels()
    plt.xticks(rotation=30, ha='right')
    plt.ylabel("Score")
    plt.tight_layout()
    plt.show()
    fig.savefig(save_pth+file_name+file_ext)
```

```python
def box_plot(X, data, file_name, y_label='score'):
    fig, ax = plt.subplots(figsize=(5, 5))
    bp = ax.boxplot(data, patch_artist=True)    
    edge_color='blue'
    fill_color='cyan'
    for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
        plt.setp(bp[element], color=edge_color)
    for patch in bp['boxes']:
        patch.set(facecolor=fill_color)   
    ax.set_xticklabels(X)
    plt.xticks(rotation=30, ha='right')
    ax.yaxis.grid(True)
    ax.xaxis.grid(False)
    plt.ylabel(y_label)
    plt.tight_layout()
    plt.show()
    fig.savefig(save_pth+file_name+file_ext)
```

```python
def box_p(X, data, file_name, y_label='score'):
    
    condition = copy.deepcopy(X)
    for i in range(1, len(condition), 2):
        condition[i] = condition[i-1]
        
    a = ['FP32', 'INT8']
    b = int(len(X)/2)
    a = [a[i % len(a)] for i in range(len(a) * b)]    
        
    df_dict = {
        "Score":np.array([]),
        "Method":np.array([]),
        "Model":np.array([])
    }
    df = pd.DataFrame(df_dict)
    
    for i in range(int(len(data))):
        results_dict = {
            "Score": data[i],
            "Method": condition[i],
            "Model":a[i]
            }
        results_df = pd.DataFrame(results_dict)
        df = pd.concat([df, results_df], ignore_index=True)
        
    ax = sns.boxplot(data=df, x="Method", y="Score", hue="Model")

    #plt.savefig(save_pth+file_name+file_ext)
```
