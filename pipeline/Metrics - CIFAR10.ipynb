```python
ce_loss_acc_path = '../../results/Crossentropy_Resnet18_loss_acc.csv'
edl_loss_acc_path = '../../results/Evidential_Resnet18_loss_acc.csv'
ce_entropy_path = '../../results/Crossentropy_Resnet18_entropy.csv'
ce_q_entropy_path = '../../results/Crossentropy_Resnet18_quant_entropy.csv'
edl_entropy_path = '../../results/Evidential_Resnet18_entropy.csv'
edl_q_entropy_path = '../../results/Evidential_Resnet18_quant_entropy.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
df = pd.read_csv(ce_loss_acc_path)
epoch_no = df['epoch_no']
ce_train_loss = df['train_loss']
ce_val_loss = df['val_loss']
ce_train_acc = df['train_acc']
ce_val_acc = df['val_acc']


df1 = pd.read_csv(edl_loss_acc_path)
edl_train_loss = df1['train_loss']
edl_val_loss = df1['val_loss']
edl_train_acc = df1['train_acc']
edl_val_acc = df1['val_acc']
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(epoch_no, ce_train_loss, label="Cross Entropy")
ax.plot(epoch_no, edl_train_loss, label="Evidential")
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_title('Training loss')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(epoch_no, ce_val_loss, label="Cross Entropy")
ax.plot(epoch_no, edl_val_loss, label="Evidential")
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.set_title('Validation loss')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(epoch_no, ce_train_acc, label="Cross Entropy")
ax.plot(epoch_no, edl_train_acc, label="Evidential")
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Training accuracy')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(epoch_no, ce_val_acc, label="Cross Entropy")
ax.plot(epoch_no, edl_val_acc, label="Evidential")
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Validation accuracy')
ax.legend()
plt.tight_layout()
plt.show()
```

# Entropy

```python
df1 = pd.read_csv(ce_entropy_path)
df2 = pd.read_csv(ce_q_entropy_path)
df3 = pd.read_csv(edl_entropy_path)
df4 = pd.read_csv(edl_q_entropy_path)
```

```python
ce_accuracy = df1.iloc[0]['accuracy']
ce_q_accuracy = df2.iloc[0]['accuracy']
edl_accuracy = df3.iloc[0]['accuracy']
edl_q_accuracy = df4.iloc[0]['accuracy']

ce_precision = df1.iloc[0]['precision']
ce_q_precision = df2.iloc[0]['precision']
edl_precision = df3.iloc[0]['precision']
edl_q_precision = df4.iloc[0]['precision']

ce_recall_score = df1.iloc[0]['recall_score']
ce_q_recall_score = df2.iloc[0]['recall_score']
edl_recall_score = df3.iloc[0]['recall_score']
edl_q_recall_score = df4.iloc[0]['recall_score']

ce_f1score = df1.iloc[0]['f1score']
ce_q_f1score = df2.iloc[0]['f1score']
edl_f1score = df3.iloc[0]['f1score']
edl_q_f1score = df4.iloc[0]['f1score']

ce = [ce_accuracy,ce_precision,ce_recall_score,ce_f1score]
ce_q = [ce_q_accuracy,ce_q_precision,ce_q_recall_score,ce_q_f1score]
edl = [edl_accuracy,edl_precision,edl_recall_score,edl_f1score]
edl_q = [edl_q_accuracy,edl_q_precision,edl_q_recall_score,edl_q_f1score]
```

```python
X = ['Accuracy','Precision','Recall','F1']
X_axis = np.arange(len(X))
width = 0.2

fig, ax = plt.subplots(figsize=(5, 5))
plt.bar(X_axis, ce, width, label = 'CE')
plt.bar(X_axis+width, ce_q, width, label = 'CE-Q')
plt.bar(X_axis+width*2, edl, width, label = 'EDL')
plt.bar(X_axis+width*3, edl_q, width, label = 'EDL-Q')
  
plt.xticks(X_axis+width*1.5, X)
plt.xlabel("Metric")
plt.ylabel("Score")
plt.title("Classification metrics")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
ce_brierscore = df1.iloc[0]['brierscore']
ce_q_brierscore = df2.iloc[0]['brierscore']
edl_brierscore = df3.iloc[0]['brierscore']
edl_q_brierscore = df4.iloc[0]['brierscore']

ce_expectedcalibrationerror = df1.iloc[0]['expectedcalibrationerror']
ce_q_expectedcalibrationerror = df2.iloc[0]['expectedcalibrationerror']
edl_expectedcalibrationerror = df3.iloc[0]['expectedcalibrationerror']
edl_q_expectedcalibrationerror = df4.iloc[0]['expectedcalibrationerror']


ce = [ce_brierscore,ce_expectedcalibrationerror]
ce_q = [ce_q_brierscore,ce_q_expectedcalibrationerror]
edl = [edl_brierscore,edl_expectedcalibrationerror]
edl_q = [edl_q_brierscore,edl_q_expectedcalibrationerror]
```

```python
X = ['Brier score','Expected calibration error']
X_axis = np.arange(len(X))
width = 0.2

fig, ax = plt.subplots(figsize=(5, 5))
plt.bar(X_axis, ce, width, label = 'CE')
plt.bar(X_axis+width, ce_q, width, label = 'CE-Q')
plt.bar(X_axis+width*2, edl, width, label = 'EDL')
plt.bar(X_axis+width*3, edl_q, width, label = 'EDL-Q')
  
plt.xticks(X_axis+width*1.5, X)
plt.xlabel("Metric")
plt.ylabel("Score")
plt.title("Uncertainty metrics")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
ce_inferencetime = df1.iloc[0]['inferencetime']
ce_q_inferencetime = df2.iloc[0]['inferencetime']
edl_inferencetime = df3.iloc[0]['inferencetime']
edl_q_inferencetime = df4.iloc[0]['inferencetime']


ce = [ce_inferencetime]
ce_q = [ce_q_inferencetime]
edl = [edl_inferencetime]
edl_q = [edl_q_inferencetime]
```

```python
X = ['Inference time']
X_axis = np.arange(len(X))
width = 0.2

fig, ax = plt.subplots(figsize=(5, 5))
plt.bar(X_axis, ce, width, label = 'CE')
plt.bar(X_axis+width, ce_q, width, label = 'CE-Q')
plt.bar(X_axis+width*2, edl, width, label = 'EDL')
plt.bar(X_axis+width*3, edl_q, width, label = 'EDL-Q')
  

ax.set_xticks([])
plt.xlabel("Model")
plt.ylabel("Time in seconds")
plt.title("Performance Evaluation metrics")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
